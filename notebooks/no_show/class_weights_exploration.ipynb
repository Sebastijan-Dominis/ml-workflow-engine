# No-show class weights

```python
import pandas as pd
from pathlib import Path

root = Path().resolve().parent.parent

def get_data(feature_path):
    y_train = pd.read_parquet(root / feature_path / "y_train.parquet")
    y_test = pd.read_parquet(root / feature_path / "y_test.parquet")

    return y_train, y_test

feature_paths = [
    "global/v1",
    "hotel/city_hotel_v1",
    "hotel/resort_hotel_v1",
]

for path in feature_paths:
    y_train, y_test= get_data(f"data/features/no_show/{path}")
    print(f"Feature path: {path}")
    print("---- TRAIN SET ----")
    print(y_train["no_show"].value_counts())
    print("---- TEST SET ----")
    print(y_test["no_show"].value_counts())
    print("\n")
```
